```python
import pandas as pd
import matplotlib.pyplot as plt

# Read Excel file into a DataFrame
df = pd.read_excel('TestCase.xlsx', header=2, index_col= 2)
df = df.iloc[:, 2:]
df.head(10)
```

```python
df.describe()
```

Вычислим общую выручку для каждого года:

```python

# Суммирование выручки по всем каналам сбыта за каждый год
total_revenue_2012 = df.loc['ИТОГО','Итого за год']
total_revenue_2013 = df.loc['ИТОГО', 'Итого за год.1']

# Вычисление процента увеличения выручки за 2013 год по сравнению с 2012 годом
increase_percentage = ((total_revenue_2013 - total_revenue_2012) / total_revenue_2012) * 100

print("Процент увеличения выручки за 2013 год по сравнению с 2012 годом:", increase_percentage, "%")

# Создание графика
years = ['2012', '2013']
total_revenue = [total_revenue_2012, total_revenue_2013]

plt.bar(years, total_revenue, color=['blue', 'green'])
plt.xlabel('Год')
plt.ylabel('Общая выручка')
plt.title("Общая выручка по годам \n Процент увеличения выручки за 2013 год по сравнению с 2012 годом:"+ str(round(increase_percentage,2))+ "%")
plt.show()
```

Разделим данные по каналам сбыта:

```python
# Выделение данных о выручке для каждого канала сбыта за 2012 и 2013 годы
rozn_2012 = df.loc['Розничные магазины', 'Итого за год']
rozn_2013 = df.loc['Розничные магазины', 'Итого за год.1']
opt_2012 = df.loc['Оптовая база', 'Итого за год']
opt_2013 = df.loc['Оптовая база', 'Итого за год.1']
chastn_2012 = df.loc['Частные распространители', 'Итого за год']
chastn_2013 = df.loc['Частные распространители', 'Итого за год.1']
```

```python
# Данные о выручке для каждого канала сбыта за 2012 и 2013 годы
channels = ['Розничные магазины', 'Оптовая база', 'Частные распространители']
revenue_2012 = [rozn_2012, opt_2012, chastn_2012]
revenue_2013 = [rozn_2013, opt_2013, chastn_2013]

# Построение графиков
plt.figure(figsize=(10, 6))

plt.plot(channels, revenue_2012, label='2012', marker='o')
plt.plot(channels, revenue_2013, label='2013', marker='o')

plt.title('Выручка по каналам сбыта за 2012 и 2013 годы')
plt.xlabel('Каналы сбыта')
plt.ylabel('Выручка')
plt.legend()
plt.grid(True)

plt.show()
```

Посчитаем долю каждого канала сбыта в общей выручке

```python
#Разделим выручку каждого канала сбыта на общую выручку за год
share_rozn_2012 = (rozn_2012 / total_revenue_2012) * 100
share_opt_2012 = (opt_2012 / total_revenue_2012) * 100
share_chastn_2012 = (chastn_2012 / total_revenue_2012) * 100

share_rozn_2013 = (rozn_2013 / total_revenue_2013) * 100
share_opt_2013 = (opt_2013 / total_revenue_2013) * 100
share_chastn_2013 = (chastn_2013 / total_revenue_2013) * 100

# Вывод результатов
print("Доля выручки по каналам сбыта в 2012 году:")
print("Розничные магазины: {:.2f}%".format(share_rozn_2012))
print("Оптовая база: {:.2f}%".format(share_opt_2012))
print("Частные распространители: {:.2f}%".format(share_chastn_2012))
print()
print("Доля выручки по каналам сбыта в 2013 году:")
print("Розничные магазины: {:.2f}%".format(share_rozn_2013))
print("Оптовая база: {:.2f}%".format(share_opt_2013))
print("Частные распространители: {:.2f}%".format(share_chastn_2013))
```

```python
# Вычисляем процентное увеличение выручки по каждому каналу сбыта с 2012 на 2013 год
increase_rozn = ((rozn_2013 - rozn_2012) / rozn_2012) * 100
increase_opt = ((opt_2013 - opt_2012) / opt_2012) * 100
increase_chastn = ((chastn_2013 - chastn_2012) / chastn_2012) * 100

# Вывод результатов
print("Процентное увеличение выручки по каналам сбыта с 2012 на 2013 год:")
print("Розничные магазины: {:.2f}%".format(increase_rozn))
print("Оптовая база: {:.2f}%".format(increase_opt))
print("Частные распространители: {:.2f}%".format(increase_chastn))

# Создаем график
plt.figure(figsize=(10, 6))

# Построение столбцов для каждого канала сбыта
plt.bar(['Розничные магазины', 'Оптовая база', 'Частные распространители'], 
        [increase_rozn, increase_opt, increase_chastn], 
        color=['blue', 'green', 'red'])

# Добавляем заголовки и метки осей
plt.title('Процентное увеличение выручки по каналам сбыта с 2012 на 2013 год')
plt.xlabel('Каналы сбыта')
plt.ylabel('Процентное увеличение (%)')

# Добавляем легенду
plt.grid(True)

# Показываем график
plt.show()
```

```python
channels = ['Розничные магазины', 'Оптовая база', 'Частные распространители']
shares_2012 = [share_rozn_2012, share_opt_2012, share_chastn_2012]
shares_2013 = [share_rozn_2013, share_opt_2013, share_chastn_2013]

# Создание графиков
plt.figure(figsize=(10, 5))

# График для 2012 года
plt.subplot(1, 2, 1)
plt.bar(channels, shares_2012, color='skyblue')
plt.title('Доля выручки по каналам сбыта в 2012 году')
plt.ylabel('Доля выручки, %')

# График для 2013 года
plt.subplot(1, 2, 2)
plt.bar(channels, shares_2013, color='salmon')
plt.title('Доля выручки по каналам сбыта в 2013 году')
plt.ylabel('Доля выручки, %')

# Отображение графиков
plt.tight_layout()
plt.show()
```

```python
# Вычисляем изменения доли выручки для каждого канала сбыта
share_changes = [shares_2013[i] - shares_2012[i] for i in range(len(channels))]

# Создаем график для изменений доли выручки
plt.figure(figsize=(8, 5))
plt.bar(channels, share_changes, color='lightgreen')
plt.title('Изменение доли выручки по каналам сбыта между 2012 и 2013 годами')
plt.xlabel('Каналы сбыта')
plt.ylabel('Изменение доли выручки, %')
plt.axhline(y=0, color='gray', linestyle='--', linewidth=0.8)  # Добавляем линию на уровне 0
plt.xticks(rotation=45)  # Поворачиваем подписи оси X для лучшей читаемости
plt.show()
```

В 2013 году доля выручки от оптовой базы увеличилась, а доли выручки от розничных магазинов и  от частных распространителей снизились.
